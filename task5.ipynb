```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales_data.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

## Data Exploration Summary
- Dataset contains 8 rows and 6 columns
- No missing values found
- Data includes Sales, Profit, Product and Region information
- Dataset is clean and ready for analysis

```python
df.groupby("Product")["Sales"].sum().sort_values(ascending=False)
```

```python
df.groupby("Region")["Sales"].sum()
```

```python
df.groupby("Product")["Profit"].sum()
```

```python
df.groupby("Product")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Product")
plt.show()
```

```python
df.groupby("Region")["Sales"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales by Region")
plt.show()
```

```python
df["Sales"].plot(kind="line")
plt.title("Sales Trend")
plt.show()
```

## Key Insights
- Laptop has highest total sales
- North region contributes major revenue
- Sales trend shows fluctuations across products
- Dataset analysis helps identify top-performing categories
